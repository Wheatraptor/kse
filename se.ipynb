```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def pendula(freq):
    def rhs(t, X):
        x, y = X
        return [y, - freq * freq * x]
    return rhs

def eq_quiver(rhs, limits, N=16):
    xlims, ylims = limits
    xs = np.linspace(xlims[0], xlims[1], N)
    ys = np.linspace(ylims[0], ylims[1], N)
    U = np.zeros((N, N))
    V = np.zeros((N, N))
    for i, y in enumerate(ys):
        for j, x in enumerate(xs):
            vfield = rhs(0., [x, y])
            u, v = vfield
            U[i][j] = u
            V[i][j] = v
    return xs, ys, U, V
```

```python
def plotOnPlane(rhs, pts, times, limits):
    plt.close()
    xlims, ylims = limits
    plt.xlim(xlims[0], xlims[1])
    plt.ylim(ylims[0], ylims[1])
    plt.axes().set_aspect('equal', adjustable='box')
    xs, ys, U, V = eq_quiver(rhs, limits, 16)
#     plt.figure(figsize=(10, 10))
    plt.quiver(xs, ys, U, V, alpha=0.5)
    
    for pt, maxT in zip(pts, times):
        sol = solve_ivp(rhs, [0, maxT], pt, method='RK45', rtol=1e-12)
        xs, ys = sol.y
        plt.plot(xs, ys, 'k-')
```

```python
freq = 1
rhs = pendula(freq)
limits = [(-2.0, 2.0), (-2.0, 2.0)]
plotOnPlane(rhs, [(0., 1.0), ], [10.], limits)
```

```python
def pependula(freq):
    def rhs(t, X):
        x, y = X
        return [y, x**3-x]
    return rhs

mas = []
for i in range(-100, 100):
    for j in range(-100, 100):
        mas.append((i/100, i/100))

freq = 1
rhs = pependula(freq)
leem = 2.0
limits = [(-leem, leem), (-leem, leem)]
plotOnPlane(rhs, mas, [10.], limits)
# mas
```
