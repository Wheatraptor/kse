```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from pylab import rcParams
```

```python
def pendula():
    def rhs(t, X):
        x, y = X
        return [y, x**4 / 4 + x**3 / 3 - x**2]
    return rhs
```

```python
def eq_quiver(rhs, limits, N=16):
    xlims, ylims = limits
    xs = np.linspace(xlims[0], xlims[1], N)
    ys = np.linspace(ylims[0], ylims[1], N)
    U = np.zeros((N, N))
    V = np.zeros((N, N))
    for i, y in enumerate(ys):
        for j, x in enumerate(xs):
            vfield = rhs(0., [x, y])
            u, v = vfield
            U[i][j] = u
            V[i][j] = v
    return xs, ys, U, V
```

```python
def plotOnPlane(rhs, pts, times, limits):
    plt.close()
    xlims, ylims = limits
    plt.figure(figsize=(20,15))
    plt.xlim(xlims[0], xlims[1])
    plt.ylim(ylims[0], ylims[1])
    xs, ys, U, V = eq_quiver(rhs, limits, 32)
    widths = np.linspace(10, 20, xs.size)
    plt.quiver(xs, ys, U, V, alpha=0.1, linewidths=widths)
    for pt, maxT in zip(pts, times):
        sol = solve_ivp(rhs, [0, maxT], pt, method='RK45', rtol=1e-12)
        xs, ys = sol.y
        plt.plot(xs, ys, 'k-')
```

```python
freq = 1
rhs = pendula()
limits = [(-5.0, 5.0), (-5.0, 5.0)]
pts = []
for i, j in enumerate(range(0, 100)):
    pts.append((i/10-5, j/10-5))
pts.extend(pts)
timespos = np.empty(100)
timespos.fill(150)
timesneg = np.empty(100)
timesneg.fill(-150)
times = np.concatenate((timespos, timesneg), axis=0)
plotOnPlane(rhs, pts, times, limits)
```
